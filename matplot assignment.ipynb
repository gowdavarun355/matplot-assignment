```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Max=np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
Min=np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
months=np.arange(12)
plt.plot(months,Max,'ro')
plt.plot(months, Min, 'bo')
plt.xlabel('Month')
plt.ylabel('Min and max temperature')

             
```

```python
from scipy import optimize
def yearly_temps(times, avg, ampl, time_offset):
    return (avg
            + ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temps, months,
                                      Max, [20, 10, 0])
res_min, cov_min = optimize.curve_fit(yearly_temps, months,
                                      Min, [-40, 20, 0])
```

```python
days = np.linspace(0, 12, num=365)

plt.figure()
plt.plot(months, Max, 'ro')
plt.plot(days, yearly_temps(days, *res_max), 'r-')
plt.plot(months, Min, 'bo')
plt.plot(days, yearly_temps(days, *res_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```

```python
import pandas as pd
url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
```

```python
titanic.head()
```

```python
freq = pd.Series(titanic['sex']).value_counts()
plt.pie(freq, labels=freq.index, autopct='%.2f%%')
plt.axis('equal')
```

```python
titanic = titanic.dropna(subset=['sex'])

mapping = {'male' : 'blue', 'female' : 'red'}
plt.scatter(titanic['fare'], titanic['age'], alpha=0.5, c=titanic['sex'].map(mapping))
```
